```python
import pandas as pd
```

```python
df_1991 = pd.read_csv('IDH_1991.csv', encoding="latin1", sep = ';')
df_2000 = pd.read_csv('IDH_2000.csv', encoding="latin1", sep = ';')
```

```python
df_1991 = df_1991.replace(',','.', regex=True)
df_2000 = df_2000.replace(',','.', regex=True)
```

```python
coluna_object = df_1991.select_dtypes(include=['object']).columns
coluna_object = coluna_object.drop(['Bairro ou grupo de bairros'])
for coluna in coluna_object:
    df_1991[coluna] = pd.to_numeric(df_1991[coluna], errors='coerce')
print(df_1991.dtypes)
```

```python
coluna_object = df_2000.select_dtypes(include=['object']).columns
coluna_object = coluna_object.drop(['Bairro ou grupo de bairros'])
for coluna in coluna_object:
    df_2000[coluna] = pd.to_numeric(df_2000[coluna], errors='coerce')
print(df_2000.dtypes)
```
